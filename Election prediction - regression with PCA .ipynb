### Prediction of Real-time Election Results: Portugal 2019 

### Variable Description

TimeElapsed (Numeric): Time (minutes) passed since the first data acquisition <br>
time (timestamp): Date and time of the data acquisition<br>
territoryName (string): Short name of the location (district or nation-wide)<br>
totalMandates (numeric): MP's elected at the moment<br>
availableMandates (numeric): MP's left to elect at the moment<br>
numParishes (numeric): Total number of parishes in this location<br>
numParishesApproved (numeric): Number of parishes approved in this location<br>
blankVotes (numeric): Number of blank votes<br>
blankVotesPercentage (numeric): Percentage of blank votes<br>
nullVotes (numeric): Number of null votes<br>
nullVotesPercentage (numeric): Percentage of null votes<br>
votersPercentage (numeric): Percentage of voters<br>
subscribedVoters (numeric): Number of subscribed voters in the location<br>
totalVoters (numeric): Percentage of blank votes<br>
pre.blankVotes (numeric): Number of blank votes (previous election)<br>
pre.blankVotesPercentage (numeric): Percentage of blank votes (previous election)<br>
pre.nullVotes (numeric): Number of null votes (previous election)<br>
pre.nullVotesPercentage (numeric): Percentage of null votes (previous election)<br>
pre.votersPercentage (numeric): Percentage of voters (previous election)<br>
pre.subscribedVoters (numeric): Number of subscribed voters in the location (previous election)<br>
pre.totalVoters (numeric): Percentage of blank votes (previous election)<br>
Party (string): Political Party<br>
Mandates (numeric): MP's elected at the moment for the party in a given district<br>
Percentage (numeric): Percentage of votes in a party<br>
validVotesPercentage (numeric): Percentage of valid votes in a party<br>
Votes (numeric): Percentage of party votes<br>
Hondt (numeric): Number of MP's according to the distribution of votes now<br>
FinalMandates (numeric): Target: final number of elected MP's in a district/national-level<br>

#### PCA Analysis


```python
#Reducing dimensionality by reducing correlated factors
cols_grp1=['totalMandates','numParishesApproved','blankVotes','nullVotes','subscribedVoters','totalVoters','pre.blankVotes','pre.nullVotes','pre.subscribedVoters','pre.totalVoters','Percentage','Mandates','pre.blankVotesPercentage','pre.votersPercentage']
cols_grp2=["Votes","Hondt"]
# Standard scaling
df1=df[cols_grp1]
df2=df[cols_grp2]
sc=StandardScaler()
a=sc.fit_transform(df1)
df_1=pd.DataFrame(a,columns=df1.columns)
b=sc.fit_transform(df2)
df_2=pd.DataFrame(b,columns=df2.columns)
```

```python
pca=PCA(n_components=1)
new_var1=pca.fit_transform(df_1)
new_var2=pca.fit_transform(df_2)
# updating dataframe 
df_new = pd.concat((df, pd.DataFrame(new_var1)), axis=1)
df_new.rename({0: 'PCA_1'}, axis=1, inplace = True)
df_new.drop(cols_grp1, axis=1, inplace=True)
df_new = df_new = pd.concat((df_new, pd.DataFrame(new_var2)), axis=1)
df_new.rename({0: 'PCA_2'}, axis=1, inplace = True)
df_new.drop(cols_grp2, axis=1, inplace=True)
df_new.head()
```

### Regression Modelling

```python
#Preparing Dataset
x = df.drop(columns=['FinalMandates'])
y = df[["FinalMandates"]]
```

```python
a=sc.fit_transform(x)
```

```python
df_x=pd.DataFrame(a,columns=x.columns)
```

```python
df_x.head()
```

```python
#Splitting the data into training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df_x,y,test_size=0.20,random_state=45)
```

```python
#defining a function to find model score,r2 score for the given dataset
model=[LinearRegression(),SVR(),DecisionTreeRegressor(),KNeighborsRegressor(),Lasso(),Ridge(),ElasticNet(),RandomForestRegressor(),AdaBoostRegressor()]

```

```python
for m in model:
    m.fit(x_train,y_train)
    print(m)
    predm=m.predict(x_test)
    print('MAE:',mean_absolute_error(y_test,predm))
    print('MSE:',mean_squared_error(y_test,predm))
    print('RMSE:',np.sqrt(mean_squared_error(y_test,predm)))
    print('R2 score:',r2_score(y_test,predm))
    print('*'*100)
    print('\n')       
```

##### Observations: 
a. Random Forest Regressor and Decision Tree Regressor Model are performing the best regression.  
b. Both models have extremely low error and high R2 score.  
c. Now we will tune to choose the best results.  

```python
#We will check for RFC and DTC individually
#Estimating parameters of DecisionTree Classifiers for best fit
dtr_rg = DecisionTreeRegressor()
parameters={'criterion':['mse'],'max_depth': np.arange(3, 15)}
clf = GridSearchCV(dtr_rg, parameters, cv=5)
clf.fit(x,y)
clf.best_params_
```

```python
dtr_rg = DecisionTreeRegressor(criterion="mse",max_depth=8)
dtr_rg.fit(x_train,y_train)
print('Score:',dtr_rg.score(x_train,y_train))
y_pred_dtr=dtr_rg.predict(x_test)
print('Mean absolute error:',mean_absolute_error(y_test,y_pred_dtr))
print('Mean squared error:',mean_squared_error(y_test,y_pred_dtr))
print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,y_pred_dtr)))
print('R2 score:',r2_score(y_test,y_pred_dtr))
```

```python
#Plotting our prediction vs test results with Decision Tree model 
plt.scatter(x=y_test,y=y_pred_dtr,marker='o',color='red',alpha=0.2)
plt.plot(x,x,'g--',alpha=0.1)
plt.xlim(0,20)
plt.ylim(0,20)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

##### Observations: 
a. We have few outliers , but most points are lying on the straight line, which shows close match of true and predicted values.  
b. Our model seems to be an accurate predictor.
